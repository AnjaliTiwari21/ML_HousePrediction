<b>Understand the problem:</b> This dataset is a csv file which consist of list of houses in Belfast. This dataset is used to predict the house prices in belfast using features(dependent and independent variables).

<b>Hypothesis Generation:</b> When you formulate a hypothesis, you actually try to achieve the results on your new experiemnt. For this you create a Null Hypothesis (Ho) and Alternate Hypothesis (Ha). 
The features you are considering which could be an influence on the response variable is what you have to achieve. For this you have Ho where there is no impact of the feature on response variable. Whereas, Ha states that there exist a direct impact of the feature on response variable.

We can either reject or fail to reject the null hypothesis in statistical analysis. It is based on setting up a confidence Interval which is this case is 95% or 0.05. The value of p decides whether to reject the null hypothesis or fail to reject the hypothesis.
1. Reject the null hypothesis : If p value < 0.05 
2. Fail to reject the null hypothesis: If p > 0.05


<h1> ANOVA </h1>

It is used to estimate the variance of continuos variable by using the categorical variables. It uses F statistic and pvalues to determine the factors which influence the target variable in our case estimated price.

```python
dataset.groupby(['garage size'])['estimated_price'].describe()
```

```python
#Storing data of categorical variables and the target variable which is continuos
data=dataset[['estimated_price','postal_code','house_type','heating','window type','garage type','garage size']]
```

```python
import scipy.stats as stats
cat=['estimated_price','postal_code','house_type','heating','window type','garage type','garage size']
pvals = []
dat=pd.DataFrame()
dat['features'] = cat
for c in cat:
    samples = []
    for cls in data[c].unique():
        s = data[data[c] == cls]['estimated_price'].values
        samples.append(s)
        pval = stats.f_oneway(*samples)[1]
    pvals.append(pval)
dat['pval']=pvals
```

```python
dat['disparity'] = np.log(1./dat['pval'].values) 
dat.sort_values('disparity',inplace=True)
sns.barplot(data=dat, x = 'features', y='disparity') 
plt.xticks(rotation=90) 
plt 
```

<b>Creating Dummy Variables</b>

```python
# Firstly, transforming the numeric features using log
dataset[['estimated_price','bedroom','bathroom']]= np.log(dataset[['estimated_price','bedroom','bathroom']])
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
cat_var=["postal_code","house_type","garage size","garage type"]
num_var=["bedroom","bathroom","driveway","parking","garden","roofspace"]
# Instantiate encoder for categorical variables and scaler for numerical variables
scaler = StandardScaler()
#ohe    = OneHotEncoder(sparse=False)
```

```python
#One Hot encoding
# Scale and Encode Separate Columns
scaled_columns  = pd.DataFrame(scaler.fit_transform(dataset[num_var]),columns=num_var)
encoded_columns =  pd.get_dummies(dataset[cat_var],columns=cat_var, prefix=cat_var)
#Joining the categorical and numerical data together
processed_data=scaled_columns.join(encoded_columns)
# Concatenate (Column-Bind) Processed Columns Back Together
#processed_data = np.concatenate([scaled_columns, encoded_columns],  axis=1)
```

```python
from sklearn.model_selection import train_test_split

x = processed_data.drop(['estimated_price'], axis=1)
y= processed_data['estimated_price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

<b> MACHINE LEARNING ALGORITHMS </b>
1. Regression : task of predicting the value of target(numerical variable) by building a model based on the one or more predictors the predictors can either be numerical or the categorical variables.

```python
from sklearn.linear_model import LinearRegression
```

```python
linear= LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
pred=linear.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2_error=r2_score(y_test, pred)
mean_error=mean_absolute_error(y_test, pred)
print("r2",r2_error)
print("mean_error",mean_error)
```

<b>LASSO REGRESSION<b>

```python
from sklearn.linear_model import Lasso

#Choosing an alpha value will make your prediction better. Hence rather than going with random value use cross validation

regr = Lasso(alpha=0.00020, max_iter=50000)
regr.fit(x_train, y_train)

# run prediction on the training set to get a rough idea of how well it does
y_pred = regr.predict(x_test)
```

```python
r2_error=r2_score(y_test, y_pred)
mean_error=mean_absolute_error(y_test, y_pred)
print("Lasso score r2",r2_error)
print("Lasso score mean_error",mean_error)
```
