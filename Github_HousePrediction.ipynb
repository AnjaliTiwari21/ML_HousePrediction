```python
#The estimated price contains the null values ,different data type values and negative values which is not possible in estimated price
#We need to work on this to work on the data for prediction

#Dropping the row having value has Nan
dataset = dataset[pd.notnull(dataset['estimated price'])] 


#Removing the rows containing negative estimated price and any other data type values.
dataset=dataset[dataset['estimated price'] >0]

dataset.tail()
```

```python
#Deleting points
dataset.sort_values(by = 'estimated price', ascending = True)[:2]
dataset = dataset.drop(dataset[dataset['estimated price'] == 89.95].index)
#dataset = dataset.drop(dataset[dataset['Id'] == 524].index)
```

```python
# Fill NaN with the median of the column
dataset['bathroom'] = dataset['bathroom'].fillna(dataset['bathroom'].median())
dataset['bedroom'] = dataset['bedroom'].fillna(dataset['bedroom'].median())
dataset['recpt'] = dataset['recpt'].fillna(dataset['recpt'].median())
```

```python
#Dropping rows where nun values are only in heating column
#dataset=dataset[pd.notnull(dataset['heating'])]
dataset.dropna(subset=['heating'],inplace=True)
```

```python
# Deleting rows where the value for heating column is null
import numpy as np
dataset.replace(to_replace =np.nan,  
                            value ="N",inplace=True) 
dataset
```

```python
#One Hot encoding
dataset = pd.get_dummies(dataset)
```

```python
from sklearn.model_selection import train_test_split

x = dataset.drop(['estimated price'], axis=1)
y= dataset['estimated price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
linear= LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
pred=linear.predict(x_test)
```

```python
r2_error=r2_score(y_test, pred)
mean_error=mean_absolute_error(y_test, pred)
print("r2",r2_error)
print("mean_error",mean_error)
```

```python
#Accuracy of the model
error=(mean_error/y_test.mean())
print(100-100*error)
```
